# Data for Covid-19 in Italy and Worldwide

Data source: WHO situation reports
Period examined: from 20th Feruary 2020

So far, the data suggest that the epidemics is spreading with a polynomial growth (3rd degree) both in Italy and worldwide. 

```python
#Reading the data
import pandas as pd

df = pd.read_csv("Covid-19_Italy_WHO_data.csv")
df["Date"] = pd.to_datetime(df['Date'], dayfirst=True)
print(df.head(5))
```

```python
#Plotting and curve fitting for confirmed cases in Italy
import matplotlib.pyplot as plt
date = df["Date"]
cases_italy = df["Total confirmed cases"]
plt.xticks(rotation=90)
plt.plot(date, cases_italy)
plt.show()

def create_date_conv(date):
    days = []
    a = list(date)
    for x in range (1, len(a)+1):
        days.append(x)
    return days
    
a = create_date_conv(date)

from numpy import *
import numpy as np
import pylab

# data to fit
x = a
y = np.array(cases_italy)
print(y)

# fit the data with a 2nd degree polynomial
z2 = polyfit(x, y, 2) 
p2 = poly1d(z2)

z3 = polyfit(x, y, 3)
p3 = poly1d(z3)

xx = linspace(1, len(x))
pylab.plot(x, y, 'o', xx, p2(xx),'-g', xx, p3(xx),'-b')
pylab.legend(['data to fit', '2nd degree polynomial', '3rd degree polynomial'])
pylab.show()
    
        

```
